# 00_introduction
# Project 1 — Market Reality & Dynamics

---

## 1. Why This Project Exists

Phần lớn nhà đầu tư bắt đầu bằng:

- Indicator
- Signal
- Backtest
- Forecast model

Nhưng rất ít người bắt đầu bằng một câu hỏi nền tảng hơn:

> Thị trường thực sự vận hành như thế nào?

Trước khi nói về Risk, Fat-tail hay Decision Theory, chúng ta phải hiểu cơ chế thực tế của thị trường.

Project này xây dựng nền móng đó.

---

## 2. Core Premise

Thị trường không ổn định không phải vì “ngẫu nhiên”.

Sự bất ổn là một đặc điểm cấu trúc.

Volatility, crash, bubble hay liquidity shock không phải tai nạn hiếm gặp.
Chúng xuất hiện từ:

- Cấu trúc thị trường
- Đòn bẩy tài chính
- Ràng buộc thanh khoản
- Hành vi con người

Hiểu được cơ chế này quan trọng hơn việc cố gắng dự đoán giá.

---

## 3. Project Scope

Project này được chia thành ba lớp chính:

---

### I. Market Structure

Tập trung vào nền tảng cơ học của thị trường.

Nội dung:

- Order flow và cơ chế hình thành giá
- Liquidity và bid–ask spread
- Vai trò của market maker
- Leverage và margin
- Forced liquidation

Câu hỏi trung tâm:

> Thanh khoản đến từ đâu và khi nào nó biến mất?

---

### II. Market Dynamics

Tập trung vào cách hệ thống vận động theo thời gian.

Nội dung:

- Volatility clustering
- Regime shift
- Momentum vs mean reversion
- Correlation breakdown

Câu hỏi trung tâm:

> Vì sao thị trường không phải là một hệ thống ổn định theo thời gian?

---

### III. Behavioral Distortion

Tập trung vào cách hành vi con người làm méo hệ thống.

Nội dung:

- Overconfidence
- Herding
- Loss aversion
- Panic selling

Câu hỏi trung tâm:

> Khi nào tâm lý tạo ra tail risk?

---

## 4. What This Project Is NOT

Project này:

- Không dự báo giá
- Không đưa ra tín hiệu giao dịch
- Không khuyến nghị đầu tư
- Không tối ưu chiến lược

Mục tiêu duy nhất là hiểu cách instability hình thành trước khi bàn về risk hay decision.
